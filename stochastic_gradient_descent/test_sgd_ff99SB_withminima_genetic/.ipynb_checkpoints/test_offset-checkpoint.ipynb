```python
%pylab inline
import matplotlib.pyplot  as plt
from matplotlib.mlab import griddata
from numpy import *
#import numpy as np
from scipy import interpolate
import mdtraj as md
import math
import sys,os
import glob
import seaborn as sbn
sbn.set_style("whitegrid")
```

```python
input_fit = loadtxt("fit_offset.dat",usecols=[0,1,2,3],skiprows=2)
n_struct = input_fit[:,0]
amber    = input_fit[:,1]
quantum  = input_fit[:,2]
amberoff = input_fit[:,3]

color=sbn.color_palette()
plt.figure(figsize=(10,10))
plt.plot(n_struct,amberoff,'x--',label='Amber+Off',markersize=2,color=color[0])
plt.plot(n_struct,amber,'o--',label='Amber',markersize=2,color=color[1])
plt.plot(n_struct,quantum,'^--',label="Quantum",markersize=2,color=color[2])
plt.legend(loc="best",fontsize=20)
plt.xlabel(r"Number of structures",fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel(r"Energy / kcal$\cdot$mol$^{-1}$",fontsize=20)
plt.yticks(fontsize=20)
plt.xlim(0,41)
#plt.ylim(-984.6,-984.5)
#plt.tight_layout()'''
```
