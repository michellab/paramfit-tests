```python
%pylab inline
import matplotlib.pyplot  as plt
from matplotlib.mlab import griddata
from numpy import *
#import numpy as np
from scipy import interpolate
import mdtraj as md
import math
import sys,os
import glob
import seaborn as sbn
sbn.set_style("whitegrid")
```

```python
f99 = loadtxt("energies_ff99.dat",usecols=[0])
f99SB = loadtxt("energies_ff99SB.dat",usecols=[0])

xaxis = linspace(1,len(f99),len(f99))
color=sbn.color_palette()
plt.figure(figsize=(10,10))
plt.plot(xaxis,f99,'o--',label='f99',markersize=2,color=color[1])
plt.plot(xaxis,f99SB,'^--',label="f99SB",markersize=2,color=color[2])
plt.legend(loc="best",fontsize=20)
plt.xlabel(r"Number of structures",fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel(r"Energy / kcal$\cdot$mol$^{-1}$",fontsize=20)
plt.yticks(fontsize=20)
plt.xlim(0,48)
#plt.ylim(-984.6,-984.5)
#plt.tight_layout()'''
```
