```python
%pylab inline
import matplotlib.pyplot  as plt
from matplotlib.mlab import griddata
from numpy import *
#import numpy as np
from scipy import interpolate
import mdtraj as md
import math
import sys,os
import glob
import seaborn as sbn
sbn.set_style("whitegrid")
```

```python
def relative(energies):
    #here I create a list with relative values of energy
    # in this way the minimum will be in 0.0
    minimum = min(energies)
    rel_energies  = []
    for val in energies:
        new_val = val - minimum
        rel_energies.append(new_val)
    return rel_energies
```

```python
input_fit = loadtxt("f99_f99SB_fit.dat",usecols=[0,1,2],skiprows=2)
n_struct = input_fit[:,0]
f99    = relative(input_fit[:,1])
f99SB  = relative(input_fit[:,2])

color=sbn.color_palette()
plt.figure(figsize=(10,10))
plt.plot(n_struct,f99,'o--',label='f99',markersize=2,color=color[1])
plt.plot(n_struct,f99SB,'^--',label="f99SB",markersize=2,color=color[2])
plt.legend(loc="best",fontsize=20)
plt.xlabel(r"Number of structures",fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel(r"Energy / kcal$\cdot$mol$^{-1}$",fontsize=20)
plt.yticks(fontsize=20)
plt.xlim(0,48)
#plt.ylim(-984.6,-984.5)
#plt.tight_layout()'''
```
