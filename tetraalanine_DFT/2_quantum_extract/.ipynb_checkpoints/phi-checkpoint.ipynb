```python
%pylab inline
import matplotlib.pyplot  as plt
from numpy import *
import mdtraj as md
import math
```

```python
import mdtraj as md
traj = md.load('03_Prod.netcdf',top='system.prmtop')
print(traj)


```

```python
def dictionary_atoms(traj):
    md_top = traj.topology
    dict_atoms = {}
    for i in range(0,43): #hard coding for the moment   
        dict_atoms[md_top.atom(i)] = i
    
    return dict_atoms
    
    
```

```python
atom_numb = dictionary_atoms(traj)
print(atom_numb)
```

```python
#psi_indices, phi_indices = [6, 8, 14, 16], [4, 6, 8, 14]
psi_indices, phi_indices = [16, 18, 24, 26], [14, 16, 18, 24]
#psi_indices, phi_indices = [26, 28, 34, 36], [24, 26, 28, 34]
```

```python
angles = md.geometry.compute_dihedrals(traj, [phi_indices, psi_indices])
```

```python
#conversion form rad to deg:
for i in range(0,len(angles)):
    angles[i][0] = angles[i][0]*57.29
    angles[i][1] = angles[i][1]*57.29

phi = []
psi = []

for i in range(0,len(angles)):
    phi.append(angles[i][0])
    psi.append(angles[i][1])
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.scatter(phi,psi,s=20,marker="o")
ax.set_xlabel("Phi $\phi$",fontsize=20)
ax.set_ylabel("Psi $\psi$",fontsize=20)
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
```

```python
hist_phi, bins_phi = histogram(phi,bins=100)
#360/100 :  3.6 deg
#360/360 :  1.0 deg
#this comes form a stackoverflow:
width =  0.7*(bins_phi[1]-bins_phi[0])
center = (bins_phi[:-1]+bins_phi[1:])/2

fig,ax = plt.subplots(figsize=(10,5))
ax.bar(center,hist_phi,align="center",width = width)
```

```python
dict_phi_fe = {}

for i in range (0, len(hist_phi)-1):
    #print(hist_phi[i])
    if hist_phi[i] == 0 :
        dg = 0 
    else:
        dg = -0.593*math.log(hist_phi[i]/50000)
        
        bin_val = bins_phi[i]
        
        dict_phi_fe[bin_val] =[hist_phi[i],dg]
    
    #print(dg)
    
dict_phi_fe
#Free energy from the most likely state :  
#DG = -kT*ln(P(i)/sum(P)) 
#dg = -0.593 * math.log(i/50000)
#print(dg)
```

```python
hist_psi, bins_psi = histogram(psi,bins=100)
#360/100 :  3.6 deg
#360/360 :  1.0 deg
#this comes form a stackoverflow:
width =  0.7*(bins_psi[1]-bins_psi[0])
center = (bins_psi[:-1]+bins_psi[1:])/2

fig,ax = plt.subplots(figsize=(10,5))
ax.bar(center,hist_psi,align="center",width = width)
```

```python
dict_psi_fe = {}

for i in range (0, len(hist_psi)-1):
    #print(hist_psi[i])
    if hist_psi[i] == 0 :
        dg = 0 
    else:
        dg = -0.593*math.log(hist_psi[i]/50000)
        
        bin_val = bins_psi[i]
        
        dict_psi_fe[bin_val] =[hist_psi[i],dg]
    
    #print(dg)
    
dict_psi_fe
#Free energy from the most likely state :  
#DG = -kT*ln(P(i)/sum(P)) 
#dg = -0.593 * math.log(i/50000)
#print(dg)
```

```python
hist_phipsi,bin_phi,bin_psi = histogram2d(phi,psi,bins=100)
#360/100 :  3.6 deg
#360/360 :  1.0 deg
#this comes form a stackoverflow:
#Flipping to have a Ramachandran appearance
hist_phipsi= rot90(hist_phipsi)
hist_phipsi = flipud(hist_phipsi)
#masked the zero values
hist_masked = ma.masked_where(hist_phipsi==0,hist_phipsi)

fig2 = plt.figure(figsize=[10,10])
plt.pcolormesh(bin_phi,bin_psi,hist_masked)
plt.xlabel("$\Phi$",fontsize=20)
plt.ylabel("$\Psi$",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

cbar=plt.colorbar()
cbar.ax.set_ylabel("prob")
plt.tight_layout()
#plt.savefig("Probabilities_distribution.pdf")
```
