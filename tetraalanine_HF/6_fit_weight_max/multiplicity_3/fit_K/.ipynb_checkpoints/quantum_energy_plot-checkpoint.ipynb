```python
%pylab inline
import matplotlib.pyplot as plt
#pylab.rcParams['figure.figsize'] = (8.0, 15.0)
from numpy import *
import math
import re
import os
import seaborn as sbn
sbn.set_style("whitegrid")
```

```python
energies = loadtxt("quantum_energy.dat",usecols=[0])
x_lin = linspace(1,len(energies),len(energies))
#these are the energies + offset
#calcualte the relative energy
minimum = min(energies)
quantum_rel = []
for val in energies:
    new_val = val - minimum
    quantum_rel.append(new_val)
    
#now plot them
fig,ax = plt.subplots(figsize=[10,5])
color = sbn.color_palette()
marker = ['o', '^']
ax.plot(x_lin,quantum_rel,color=color[0],marker=marker[0],label="Quantum Relative")
plt.xlim(1,len(energies))
plt.xlabel("Structures",fontsize=20)
plt.ylabel("Energies / H",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc="best",fontsize=20)
plt.tight_layout()

```

```python
#now create a weight_quantuM_energy file
#Rules:
# val > 0.020    weight = 0 
# val < 0.020  and  val > 0.010  weight = 2
# val < 0.010  and val > 0.005   weight = 5
# val < 0.005                    weight = 10

output = open("quantum_weight.dat","w")

for i,val in enumerate(quantum_rel,0):
    if i==len(energies)-1:
        if val > 0.020 :
            output.write("%.8f     %.2f" % (energies[i],0.00))
        elif val<0.005:
            output.write("%.8f     %.2f" % (energies[i],10.00))
        else:
            output.write("%.8f" % (energies[i]))
    else:
        if val > 0.020 :
            output.write("%.8f     %.2f\n" % (energies[i],0.00))
        elif val<0.005:
            output.write("%.8f     %.2f\n" % (energies[i],10.00))
        else:
            output.write("%.8f\n" % (energies[i]))
            
output.close()
        
        
```
